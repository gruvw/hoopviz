# Games by player & season

```python
from pathlib import Path
import pandas as pd

DATA_DIR = Path("../../data/nba_database")

games_path = DATA_DIR / "Games.csv"
player_stats_path = DATA_DIR / "PlayerStatistics.csv"

games = pd.read_csv(games_path) # game metadata like gameId, date, season, teams
player_stats = pd.read_csv(player_stats_path) # player-game rows, one row per player per game

print("Games:", games.shape)
print("PlayerStatistics:", player_stats.shape)
```

```python
print(list(games.columns))
print(list(player_stats.columns))
```

```python
player_stats["gameDateTimeEst"] = pd.to_datetime(
    player_stats["gameDateTimeEst"]
)

games["gameDateTimeEst"] = pd.to_datetime(
    games["gameDateTimeEst"]
)

games.head()
```

```python
def get_season(date):
    year = date.year
    return year if date.month >= 9 else year - 1

player_stats["season"] = player_stats["gameDateTimeEst"].apply(get_season)

player_stats.head()
```

```python
games_small = games[[
    "gameId",

    "homeScore",
    "awayScore",

    # "arenaName", # map chart could be added, but probably in teams dashboard
    # "arenaCity",
    # "arenaState",

    "attendance" # can be used for another stat
]]

games_small.head()
```

```python
player_games = player_stats.merge(
    games_small,
    on="gameId",
    how="left"
)

player_games = player_games.drop(
    columns=["firstName", "lastName"],
    errors="ignore"
)

player_games = player_games[[
    "personId",
    "season",
    "gameId",
    "gameDateTimeEst",

    "playerteamCity",
    "playerteamName",
    "opponentteamCity",
    "opponentteamName",
    "gameType",
    "home",
    "win",

    "numMinutes",
    "points",
    "assists",
    "reboundsTotal",
    "reboundsDefensive",
    "reboundsOffensive",
    "blocks",
    "steals",
    "turnovers",
    "foulsPersonal",
    "plusMinusPoints",

    "fieldGoalsAttempted",
    "fieldGoalsMade",
    "fieldGoalsPercentage",
    "threePointersAttempted",
    "threePointersMade",
    "threePointersPercentage",
    "freeThrowsAttempted",
    "freeThrowsMade",
    "freeThrowsPercentage",

    "homeScore",
    "awayScore",
    "attendance"
]]

player_games = player_games.sort_values(
    ["personId", "season", "gameDateTimeEst"]
).reset_index(drop=True)

player_games.head()
```

```python
int_cols = [
    "personId",
    "season",
    "gameId",
    "home",
    "win",
    "points",
    "assists",
    "reboundsDefensive",
    "reboundsOffensive",
    "reboundsTotal",
    "blocks",
    "steals",
    "turnovers",
    "foulsPersonal",
    "plusMinusPoints",
    "fieldGoalsAttempted",
    "fieldGoalsMade",
    "threePointersAttempted",
    "threePointersMade",
    "freeThrowsAttempted",
    "freeThrowsMade",
    "homeScore",
    "awayScore",
    "attendance"
]

for col in int_cols:
    player_games[col] = (
        player_games[col]
        .fillna(0)
        .astype("Int64")
    )
```

```python
def minutes_to_float(x):
    if pd.isna(x):
        return 0

    if isinstance(x, str) and ":" in x:
        mins, secs = x.split(":")
        return int(mins) + int(secs) / 60

    return float(x)

player_games["numMinutes"] = (
    player_games["numMinutes"] # for old data format
    .apply(minutes_to_float)
)
```

```python
player_season_games = (
    player_games
    .groupby(["personId", "season"])
    .apply(
        lambda df: df.drop(
            columns=["personId", "season"]
        ).to_dict("records")
    )
    .reset_index(name="games")
)

player_season_games.head()
```

```python
import os

output_dir = "../../src/data/games_by_player"
os.makedirs(output_dir, exist_ok=True)

for person_id, df in player_games.groupby("personId"):
    df = df.sort_values(["season", "gameDateTimeEst"])

    df.to_csv(
        f"{output_dir}/{person_id}.csv",
        index=False
    )
```
